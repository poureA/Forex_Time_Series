# Step4

```python
#Seperating train and test sets.
#I have considered 95% of total data frame for train set.
idx = (len(new_df)*int(input('Enter train size in percent:')))//100
train = new_df['BC'].to_numpy()[:idx]
test = new_df['BC'].to_numpy()[idx:]
```

```python
print(f'There are {len(train)} samples in train set and {len(test)} in test set.')
```

# Step5

```python
from tensorflow.data import Dataset
from tensorflow.keras import layers , models
from tensorflow.keras.metrics import mean_squared_error , mean_absolute_error
from tensorflow.keras.callbacks import Callback
from numpy import mean
```

Since prices are submitted every minute, So I've made a decision to make a set for each 30 minutes.

```python
window_size = 30
shuffl_size = 60
batch_size = 15
```

```python
train_set = Dataset.from_tensor_slices(train)
train_set = train_set.window(window_size,shift=1,drop_remainder=True)
train_set = train_set.flat_map(lambda x : x.batch(window_size))
train_set = train_set.shuffle(shuffl_size).map(lambda x : (x[:-1],x[-1]))
train_set = train_set.batch(batch_size).prefetch(1)
```

```python
test_set = Dataset.from_tensor_slices(new_df.BC)
test_set = test_set.window(window_size,shift=1,drop_remainder=True)
test_set = test_set.flat_map(lambda x : x.batch(window_size))
test_set = test_set.batch(batch_size).prefetch(1)
```

```python
model = models.Sequential([layers.Dense(4,activation='relu',input_shape=[None,1]),
                           layers.Dense(4,activation='relu'),
                           layers.Dense(1)])
model.compile(optimizer='rmsprop',
              loss='Huber',
              metrics=['mae'])
```

```python
history = model.fit(train_set,epochs=10)
```

```python
#Visualize models performance.
epoch = range(1,len(history.epoch)+1)
results = history.history
plot(epoch,results['mae'],'green')
xlabel('Epochs')
ylabel('mae')
title('Forex EURUSD in 2020')
show()
plot(epoch,results['loss'],'orange')
xlabel('Epochs')
ylabel('Loss')
title('Forex EURUSD in 2020')
show()
```

```python
pred = model.predict(test_set)
```

```python
new_pred = pred[idx-30:-1]
```

```python
new_pred = [mean(i.flatten()) for i in new_pred]
```

```python
message = 'mse : %f , mae : %f'%(mean_squared_error(test,new_pred).numpy(),mean_absolute_error(test,new_pred).numpy())
print(message)
```

```python
#Visualize models performance on predicting.
figure(figsize=(20,10))
plot(range(len(test)),test)
plot(range(len(test)),new_pred)
ylabel('Price')
legend(['Actual price','Predicted price'])
title('Forex EURUSD in 2020')
show()
```
